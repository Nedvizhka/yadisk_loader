```python
def get_city_house_cnt(engine):
    city_house_cnt_query = f'''SELECT c.name as city, count(*) as house_cnt
                    from jkh_houses jh 
                    join city c 
                    on c.id = jh.city_id 
                    GROUP by jh.city_id
                    '''
    try:
        con_obj = engine.connect()
        city_house_cnt_db = pd.read_sql(text(city_house_cnt_query), con=con_obj)
        con_obj.close()
        exc_code = None
    except:
        city_house_cnt_db = None
        exc_code = exc.code
    return city_house_cnt_db, exc_code
```

```python
files_parse = os.listdir('saved_shit/')
```

```python
files_parse_avito = [i for i in files_parse if 'циан' not in i]
files_parse_cian = [i for i in files_parse if 'циан' in i]
```

```python
files_parse_df = pd.DataFrame(columns=['f_name', 'f_date', 'source'])
```

```python
for i in files_parse_avito:
    files_parse_df.loc[len(files_parse_df)] = [i, None, 'avito']
```

```python
for i in files_parse_cian:
    files_parse_df.loc[len(files_parse_df)] = [i, None, 'cian']
```

```python
files_parse_df.f_date = files_parse_df.f_name.apply(lambda x: datetime.strptime(x[-25:-17], '%d-%m-%y'))
```

```python
def get_source(rep_dt, source):
    if source == 'avito':
        source_name = files_parse_df[(files_parse_df.f_date == rep_dt) & 
                                     (files_parse_df.source == source)].f_name.values[0]
        
        city_df, source_df, exc_code = get_tables_info(sql_engine)
        city_df_dict = city_df.set_index('id').to_dict()['url_avito']
        
        df_s = pd.read_csv(f'saved_shit/{source_name}', delimiter=';')
        
        df_s['ad_id'] = df_s['Ссылка'].apply(lambda x: x.split('_')[-1])
        df_s['ad_id'] = df_s['ad_id'].astype('int64')
        df_s['city'] = df_s['Ссылка'].apply(lambda x: x.split('/')[3])
        df_s['city_id'] = df_s['city'].apply(lambda x: city_id_from_link(x, city_df_dict))
        df_s['addr'] = df_s.apply(lambda row: avito_addr_from_row(row, city_df), axis=1)
        df_s['city_ru'] = df_s.addr.apply(lambda x: x.split('; ')[0])
        
        return df_s
        
    else:
        source_name = files_parse_df[(files_parse_df.f_date == rep_dt) & 
                                     (files_parse_df.source == source)].f_name.values[0]
        
        city_df, source_df, exc_code = get_tables_info(sql_engine)
        city_df_dict = city_df.set_index('id').to_dict()['name']
        
        df_c = pd.read_csv(f'saved_shit/{source_name}', delimiter=',')
        
        df_c['ad_id'] = df_c['Ссылка'].apply(lambda x: x.split('/')[-2])
        df_c['ad_id'] = df_c['ad_id'].astype('int64')
        df_c['city'] = df_c['Адрес'].apply(lambda x: x.split(';')[1])
        df_c['city_id'] = df_c['city'].apply(lambda x: city_id_from_link(x, city_df_dict))
        df_c['addr'] = df_c['Адрес']
        
        return df_c
```

# avito rep

```python
from tqdm import tqdm
```

```python
rep_table = pd.DataFrame(columns=['dt', 'city', 'source', 'house_cnt', 'ad_cnt', 'ad_new_cnt', 
                                  'addr_new_cnt', 'ddt_cnt', 'qc_0_cnt' , 'qc_not_0_cnt', 'fias_none_cnt'])
```

```python
source = 'avito'
for av_file in tqdm(files_avito, total=len(files_avito)):
    file_date = datetime.strptime(av_file[21:31], '%Y_%m_%d')
    df = pd.read_csv(f'saved_data/{av_file}', encoding='cp1251', index_col=0)
    df['city'] = df.addr.apply(lambda x: x.split('; ')[0])
    
    city_list = df.city.unique().tolist()
    
    print('get source', file_date)
    df_source = get_source(file_date, source)
    
    for city_name in city_list:
        city = city_name
        house_cnt = city_cnt_df[city_cnt_df.city==city_name]['house_cnt'].values[0]
        ad_cnt = len(df_source[df_source.city_ru == city].ad_id.unique())
        
        ad_new_cnt = len(df[df.city == city_name])
        addr_new_cnt = len(df[df.city == city_name].addr.unique())
        ddt_cnt = len(df[df.city == city_name].addr.unique()) # в dadata result сохраняются только успешные запросы 
        
        df_unique = df.drop_duplicates(subset='addr')
        
        qc_0_cnt = len(df_unique[(df_unique.city == city_name) & (df_unique.qc == 0.0)])
        qc_not_0_cnt = len(df_unique[(df_unique.city == city_name) & (df_unique.qc != 0.0)])
        fias_none_cnt = len(df_unique[(df_unique.city == city_name) & (df_unique.house_fias_id.isna())])
        
        to_append = [file_date, city, source, house_cnt, ad_cnt, ad_new_cnt, addr_new_cnt, 
                     ddt_cnt, qc_0_cnt, qc_not_0_cnt, fias_none_cnt]
        rep_table.loc[len(rep_table)] = to_append
```
